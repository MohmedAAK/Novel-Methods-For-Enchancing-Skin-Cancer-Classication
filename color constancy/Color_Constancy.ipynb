```
from r
from PIL import Image
import math
import cv2
import matplotlib.pyplot as plt
im = Image.open("/content/drive/My Drive/ISIC2018_Task3_Test_Input/ISIC_0034524.jpg")
img = cv2.imread('/content/drive/My Drive/ensemblev1/model_nv_others/model1/validate/nv/ISIC_0024319.jpg')
#/content/drive/My Drive/ensemblev1/model_nv_others/model1/validate/nv/ISIC_0024363.jpg
#/content/drive/My Drive/ensemblev1/model_nv_others/model1/validate/nv/ISIC_0024479.jpg
plt.imshow(im)                           
plt.axis('off')
plt.show()
r=0
b=0
g=0
print(img[1][1])
print(img[1][1][0]*(1/(math.sqrt(3)*0.914)))
print(img[1][1][1]*(1/(math.sqrt(3)*0.304)))
print(img[1][1][2]*(1/(math.sqrt(3)*0.267)))
print(img.shape)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    img[i][j][0]=img[i][j][0]*(1/(math.sqrt(3)*0.8413510530137981))
    img[i][j][1]=img[i][j][1]*(1/(math.sqrt(3)*0.6320730283224402))
    img[i][j][2]=img[i][j][2]*(1/(math.sqrt(3)*0.3658871895424836))
    r+=img[i][j][0]
    b+=img[i][j][1]
    g+=img[i][j][2]
print((r/(450*600))/255)   
print((b/(450*600))/255)  
print((g/(450*600))/255)    
im_pil = Image.fromarray(img)  
plt.imshow(img)                           
plt.axis('off')
plt.show()  
    
```

```
import colorcorrect.algorithm as cca
img = Image.open('/content/drive/My Drive/Skin_cancer_all/skin_cancer/akiec/ISIC_0024800.jpg')
from colorcorrect.util import from_pil, to_pil
plt.imshow(img)                           
plt.axis('off')
plt.show()  
ims=to_pil(cca.automatic_color_equalization(from_pil(img)))
plt.imshow(ims)                           
plt.axis('off')
plt.show()
imss=to_pil(cca.stretch(from_pil(img)))
plt.imshow(imss)                           
plt.axis('off')
plt.show()  
imss=to_pil(cca.stretch_pre(from_pil(img)))
plt.imshow(imss)                           
plt.axis('off')
plt.show() 
imss=to_pil(cca.standard_deviation_and_luminance_weighted_gray_world(from_pil(img)))
plt.imshow(imss)                           
plt.axis('off')
plt.show()  
imss=to_pil(cca.retinex(from_pil(img)))
plt.imshow(imss)                           
plt.axis('off')
plt.show()  
```
